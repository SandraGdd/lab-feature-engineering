```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Check for missing values**

```python
#your code here

spaceship.isnull().sum()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
#your code here

spaceship = spaceship.dropna()
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
#your code here

spaceship['Cabin']= spaceship['Cabin'].str[0]
```

- Drop PassengerId and Name

```python
#your code here

spaceship= spaceship.drop(['PassengerId','Name'], axis=1)
```

- For non-numerical columns, do dummies.

```python
#your code here

spaceship_cat= spaceship.select_dtypes(include=['object'])
```

```python
spaceship_cat= pd.get_dummies(spaceship_cat, dtype = int)
```

```python
spaceship_cat
```

```python
spaceship
```

```python
spaceship_combined = pd.concat([spaceship, spaceship_cat], axis=1)
spaceship_combined
```

**Perform Train Test Split**

```python
#your code here

features = spaceship.drop('Transported', axis=1)
target = spaceship["Transported"]
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=0)
X_train.head()
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
#your code here

from sklearn.neighbors import KNeighborsClassifier   
knn = KNeighborsClassifier(n_neighbors=5) 
pred = knn.predict(X_test)        # test data 
pred
```
